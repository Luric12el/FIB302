```python
# Caso o usuário seja meio burro e escreva uma palavra pra responder a pergunta do código acima, também haverá um erro. Para corrigir, pode-se usar o try/except e o "while True" para repetir a pergunta.
while True: #prende o usuário em um loop
    try:
        ad = float(input("Quantos litros de água você adicionou ao balde? "))
        
        break # se o código chegar aqui sem erros, quebra o loop

    except ValueError: # se der o erro ValueError, cai nesse código aqui e volta para o início do loop
        print("Tem que digitar um número, bocó") 

vltotal = 10 + ad
print(f"O volume de água total no balde é de {vltotal} litros.")
```

Q3 -  Escreva um programa que calcule a área de um retângulo. Solicite ao usuário o comprimento e a largura e imprima o resultado.

```python
# Usando as mesmas lógicas vistas acima
while True: 
    try:
        base = float(input("Digite o tamanho da base do retângulo (m): ")) #pega o input da base e converte pra número (float)
        altura = float(input("Digite o valor da altura do retângulo (m): ")) # pega o input da altura e converte pra número

        break #finaliza o loop se tudo der certo
    except ValueError: # Caso ocorra ValueError (a pessoa digitar uma palavra)
        print("Por gentileza, somente números")
area = base * altura # Calcula a área do retângulo
print(f"A área do seu retângulo é {area:.2f}m²")
```

1 - Suponha que a posição de um ponto no espaço bidimensional é dado em coordenadas polares $r$, $\theta$  e queremos convertê-lo em coordenadas cartesianas x , y. Escreva um programa para fazer essa conversão. Para isso faça:

- Obter do usuário os valores de $r$ e $\theta$ 
- converter esses valores em coordenadas cartesianas usando as fórmulas: $x = r\cos{\theta}$  e $y = r\sin{\theta}$

```python
r = float(input("Digite a coordenada 'r' do seu ponto: "))
theta = float(input("Digite a coordenada 'theta' do seu ponto, em radianos: "))
x = r * math.cos(theta)
y = r * math.sin(theta)
print(f"As coordenadas do seu ponto, no sistema cartesiano, são: x = {x}; y = {y}")
```

2 - Escreva um programa que compute os principais parametros de um lançamento obliquo dado o módulo da velocidade inicial e o angulo de lançamento. Imprima a saída de maneira que os resultados sejam descritos claramente.

```python
g = 9.80665

print("CALCULADORA DE LANÇAMENTO OBLÍQUO") 

v0 = float(input("Digite a velocidade incial (m/s): "))
theta = float(input("Digite o ângulo de lançamento (rad): "))

t = (2* v0 * math.sin(theta))/g #calcula o tempo de voo
h = ((v0 * math.sin(theta))**2)/g #calcula a altura máxima
d = (v0**2 * math.sin(2*theta))/g #calcula a distância em que irá atingir o solo, considerando um plano perfeito e resistências. 

print(f"Tempo de voo: {t:.3f}s")
print(f"Altura máxima: {h:.3f}m")
print(f"Disância percorrida: {d:.3f}m")
```
